```python
import sqlite3


def sqlite_to_dict(path,table_name):
    # Connect to the SQLite database file
    conn = sqlite3.connect(path)

    # Create a cursor object
    cursor = conn.cursor()

    # Specify the table name for which you want to fetch data

    # Get all column names
    cursor.execute(f"PRAGMA table_info({table_name})")
    column_info = cursor.fetchall()
    column_names = [col[1] for col in column_info]

    # Fetch data for each column and store in a dictionary
    data_dict = {}
    for column in column_names:
        cursor.execute(f"SELECT {column} FROM {table_name}")
        column_data = cursor.fetchall()
        data_dict[column] = [row[0] for row in column_data]

    # Close the connection
    conn.close()

    return data_dict
```

```python


path = 'data/result_database_tashi.sqlite'
table_name = 'annotations'

data_dict = sqlite_to_dict(path,table_name)
# Print the data dictionary
print("Data dictionary:")
for column, data in data_dict.items():
    print(f"{column}: {len(data)}")
```
